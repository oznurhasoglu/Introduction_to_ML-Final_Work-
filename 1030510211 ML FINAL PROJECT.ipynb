# 1. VERİ DÜZENLEME İŞLEMLERİ

```python
#KÜTÜPHANELERİ TANIMLIYORUM
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
#VERİLERİMİ YÜKLÜYORUM VE KONTROL EDİYORUM
veri = pd.read_csv('data.csv')
veri
```

```python
#VERİLER İÇİNDE NULL DEĞER OLUP OLMADIĞINI KONTROL EDİYORUM.
veri.isna().sum()
```

```python
#ETİKET VERİLERİMİ YÜKLÜYORUM.
label = pd.read_csv('labels.csv')
label
```

```python
#X KOLONLARIMI BELİRLİYORUM.
X = veri.iloc[:, 1:]
X
```

```python
#Y KOLONUMU BELİRLİYORUM.
Y= label.iloc[:, 1:]
Y
```

```python
#DataConversionWarning UYARISINDAN KURTULMAK İÇİN Y KOLONUNU TEK BOYUTLU DİZİYE DÖNÜŞTÜRÜYORUM.
Y = np.ravel(Y)
#disease_type YANİ Y KOLONUMU SAYISALLAŞTIRIYORUM. 
Y = preprocessing.LabelEncoder().fit_transform(Y)
Y
```

# 2. MULTILAYER PERCEPTRON (MLP) MODEL OLUŞTURMA

```python
#KÜTÜPHANELERİ TANIMLIYORUM
import tensorflow as tf
from tensorflow import keras
```

DEĞİŞKEN DEĞERLER
-gizli katman  sayısı: 4   
-nöron sayıları: 1024,512,256,4   
-aktivasyon fonksiyonu: relu,softmax             
-optimizer fonksiyonu: adam

```python
num_classes = 4 #sınıf sayımızı belirttim.

#MODELİ OLUŞTURUYORUM
model = keras.Sequential([
    keras.layers.Dense(1024, activation='relu', input_shape=(1836,)),
    keras.layers.Dropout(0.5),  # Dropout ekledik
    keras.layers.BatchNormalization(),  # Batch Normalization ekledik
    keras.layers.Dense(512, activation='relu'),
    keras.layers.Dropout(0.5),  # Dropout ekledik
    keras.layers.BatchNormalization(),  # Batch Normalization ekledik
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dropout(0.5),  # Dropout ekledik
    keras.layers.BatchNormalization(),  # Batch Normalization ekledik
    keras.layers.Dense(num_classes, activation='softmax')
])
```

```python
#MODELİ DERLİYORUM
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

# 3. MODEL EĞİTİMİ VE DEĞERLENDİRMESİ

```python
#KÜTÜPHANELERİ TANIMLIYORUM
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import precision_score, recall_score, fbeta_score
from sklearn.metrics import classification_report
```

```python
#ÇAPRAZ DOĞRULAMA İÇİN STRATIFIEDKFOLD KULLANIYORUM
skf = StratifiedKFold(n_splits=5, shuffle=True)
```

```python
#Precision, recall ve F2 skorlarını saklamak için boş listeler oluşturuyorum
precision_scores = []
recall_scores = []
f2_scores = []
```

```python
X = X.to_numpy()
#Y = Y.to_numpy()
```

DEĞİŞKEN DEĞERLER                   
-epoch: 1000                           
-batch_size: 32                                
-average: None

```python
#ÇAPRAZ DOĞRULAMA DÖNGÜSÜ
for train_index, test_index in skf.split(X, Y): 
    #Test ve train olarak verileri rastgele bölüyorum
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = Y[train_index], Y[test_index]


    #Modeli eğitiyorum
    model.fit(X_train, Y_train, batch_size=32, epochs=1000)

    #Test verisi üzerinde tahmin yapıyorum
    y_pred_prob = model.predict(X_test)
    y_pred = np.argmax(y_pred_prob, axis=1)

    #Precision, recall ve F2 skorlarını hesaplayıp listeye ekliyorum
    precision = precision_score(Y_test, y_pred, average=None)
    recall = recall_score(Y_test, y_pred, average=None)
    f2score = fbeta_score(Y_test, y_pred, beta=2, average=None)

    precision_scores.append(precision)
    recall_scores.append(recall)
    f2_scores.append(f2score)
```

```python
# Her bir sınıf için precision, recall ve f2 score değerlerini ekrana yazdırdım
for i in range(4):
    print(f"{i}. sınıf için Precision: {np.mean([precision_scores[j][i] for j in range(5)])}")
    print(f"{i}. sınıf için Recall: {np.mean([recall_scores[j][i] for j in range(5)])}")
    print(f"{i}. sınıf için F2 Score: {np.mean([f2_scores[j][i] for j in range(5)])}")
```

```python
#PRECISION, RECALL ve F2 SKORLARININ ORTALAMASINI YAZDIRIYORUM
print("Precision:", np.mean(precision_scores))
print("Recall:", np.mean(recall_scores))
print("F2 Score:", np.mean(f2_scores))
```
